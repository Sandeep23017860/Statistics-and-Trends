```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('car_prices.csv')

# Display the first few rows of the dataset
print(df.head())
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)

# Handle missing values
df.dropna(inplace=True)
```

```python
# Descriptive statistics
descriptive_stats = df.describe()
print("Descriptive Statistics:\n", descriptive_stats)
```

```python
non_numeric_cols = df.select_dtypes(exclude=['number']).columns
print("Non-numeric columns:", non_numeric_cols)

# Handle non-numeric values
# For example, you can encode categorical variables or handle missing values
# For simplicity, we'll just drop non-numeric columns in this example
df_numeric = df.select_dtypes(include=['number'])

# Calculate correlation matrix
correlation_matrix = df_numeric.corr()

# Display correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
# Create a histogram of a numerical variable
plt.hist(df['sellingprice'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Data')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a line graph of trends over time
plt.plot(df['year'], df['sellingprice'], color='orange')
plt.title('Trends Over Time')
plt.xlabel('Year')
plt.ylabel('sellingprice')
plt.show()
```

```python
# Create a correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
